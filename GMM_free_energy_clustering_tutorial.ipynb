```python
import numpy as np
import GMM_FE

data = np.loadtxt('data/CaM_PCA_CA_contacts_REST.txt')
```

```python
gmm_FE = GMM_FE.free_energy(data, min_n_components=3, max_n_components=15, temperature=300.0, n_iterations=3, n_grids=30)
```

```python
# Estimating free energy landscape with GMM and cross-validation.
coords, FE_landscape, FE_points = gmm_FE.landscape()
```

```python
# Visualize free energy landscape
gmm_FE.visualize(savefig=False,vmax=5)
```

```python
# Cluster landscape and identify intermediate states
labels, cluster_centers = gmm_FE.cluster(coords,FE_points,data)
```

```python
# Visualize free energy landscape with cluster labels
gmm_FE.visualize(savefig=True,vmax=5)
```
